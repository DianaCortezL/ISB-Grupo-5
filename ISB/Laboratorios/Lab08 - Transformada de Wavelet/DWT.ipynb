```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generar una seal senoidal con cambio abrupto de frecuencia
t1 = np.linspace(0, 1, 1000, endpoint=False)
t2 = np.linspace(1, 2, 1000, endpoint=False)
x1 = np.sin(2 * np.pi * 5 * t1)  # Frecuencia de 5 Hz
x2 = np.sin(2 * np.pi * 20 * t2)  # Frecuencia de 20 Hz
x = np.concatenate((x1, x2))

# Aplicar la DWT
niveles = 5
coeficientes = pywt.wavedec(x, 'db4', level=niveles)

# Graficar la seal original y los coeficientes
plt.figure(figsize=(12, 12))

# Graficar la seal original primero
plt.subplot(niveles + 2, 1, 1)
plt.plot(np.concatenate((t1, t2)), x)
plt.title('Seal Original')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

# Graficar los coeficientes de detalle (invertidos)
for i, detalle in enumerate(reversed(coeficientes[1:]), 1):
    plt.subplot(niveles + 2, 1, i + 1)
    plt.plot(detalle)
    plt.title(f'Coeficientes de Detalle Nivel {i}')
    plt.ylabel('Amplitud')

# Graficar los coeficientes de aproximacin al final
plt.subplot(niveles + 2, 1, niveles + 2)
plt.plot(coeficientes[0])
plt.title(f'Coeficientes de Aproximacin ')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```
