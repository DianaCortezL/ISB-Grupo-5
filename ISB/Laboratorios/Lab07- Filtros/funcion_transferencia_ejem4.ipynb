```python
import control as ct
```

```python
# Create a variable 's' to allow algebra operations for SISO systems
s = ct.tf('s')
G  = (s + 1)/(s**2 + 2*s + 1)
```

```python
print(G)
```

```python
sys1 = ct.tf(G.num, G.den, name='sys1')
response = ct.frequency_response([sys1])

ct.bode_plot(response, initial_phase=0)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 2))
plt.text(0.5, 0.5, f'${G}$', fontsize=20, ha='center')
plt.axis('off')
plt.show()
```

```python
import control as ct

# Define the discrete-time transfer function
z = ct.tf('z')  # Define z as a transfer function
Gd = (z + 1) / (z**2 + 2*z + 1)

Gd
```

```python
print(Gd)
```

```python
from IPython.display import display, Math

display(Math(f'G_d(z) = {Gd}'))
```

```python
# Mostrar la función de transferencia
plt.text(0.1, 0.5, f'$G_d(z) = \\frac{{z+1}}{{z^2+2z+1}}$', fontsize=20)
plt.axis('off')
plt.show()
```
