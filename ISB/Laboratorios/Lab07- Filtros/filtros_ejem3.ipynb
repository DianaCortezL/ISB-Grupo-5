```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal
import neurokit2 as nk
```

```python
fs = 200
sig1 = nk.signal_simulate(duration=2, frequency=10, noise=0.1, sampling_rate=fs)

nk.signal_plot(sig1)
```

```python
plt.magnitude_spectrum(sig1, Fs=fs)
plt.grid(ls=":")
plt.show()
```

```python
sig1_f = nk.signal_filter(sig1,sampling_rate=fs, lowcut=9, highcut=11, method="butterworth", order=2)

nk.signal_plot(sig1_f)
```

```python
plt.magnitude_spectrum(sig1_f, Fs=fs)
plt.grid(ls=":")
plt.show()
```

```python
f, t, wvd = nk.signal_timefrequency(sig1,
                                    sampling_rate=fs,
                                    max_frequency=20,
                                    method="cwt",
                                    show=True)
```

```python
f, t, wvd = nk.signal_timefrequency(sig1_f,
                                    sampling_rate=fs,
                                    max_frequency=20,
                                    method="cwt",
                                    )
plt.axis("off")
plt.title("")
plt.show()
```

```python
# CWTM Method
f, t, cwtm = nk.signal_timefrequency(sig1_f,
                                         sampling_rate=fs,
                                         max_frequency=20,
                                         method="cwt",
                                         show=True)
```

```python
plt.pcolormesh(t, f, np.abs(cwtm), shading='auto')  # shading='auto' para evitar advertencias
plt.axis('off')  # Desactivar los ejes si quieres que solo se vea el espectrograma sin etiquetas

# Mostrar la gráfica
plt.show()
```
