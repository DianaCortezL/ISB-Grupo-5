```python
from scipy.signal import dlti, tf2zpk
import numpy as np
import matplotlib.pyplot as plt
```

# Sistema de Primer Orden
Este es un sistema LTI simple de primer orden. Su ecuación en diferencias es:

image.png

$$ H(z) = \frac{z}{z-0.5} \hspace{1cm} dt=0.005$$

```python
num = [1, 0]
den = [1, -0.5,1]
dt = 0.005

sys1 = dlti(num, den, dt=dt)
sys1
```

```python
# @title Polos y ceros
import zplane as zp

zp.pz(sys1)
```

```python
# @title Bode
import control as ctrl

sys1_ = ctrl.TransferFunction(num, den, dt)
ctrl.bode_plot(sys1_, Hz=True)
plt.title('Sistema 1 - Diagrama de Bode')
```

```python
zp.impulse(sys1)
```

```python
t, y = ctrl.step_response(sys1_)
plt.plot(t, y)

plt.title('Sistema 1 - Respuesta al Escalón')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(ls=":")
plt.tight_layout()
plt.show()
```

```python
# Diagrama de Nyquist
ctrl.nyquist_plot(sys1_)
plt.title('Sistema 1 - Diagrama de Nyquist')
```

```python
ctrl.pole_zero_plot(sys1_)
plt.title('Sistema 1 - Polos y Ceros')
plt.grid(ls=":")
plt.show()
```
