```python
import neurokit2 as nk
import ecg_plot
import matplotlib.pyplot as plt
```

## Paso 3: Generar Señal Sintética de ECG

```python
# Generar un ECG sintético
ecg_signal = nk.ecg_simulate(duration=10, sampling_rate=500)  # duración en segundos y frecuencia de muestreo
```

```python
ecg_signal
plt.plot(ecg_signal)
```

```python
ecg_signal, list(map(lambda x: float(x), ecg_signal))
```

```python
ecg_signal_aux = list(map(lambda x: float(x), ecg_signal))
```

```python
ecg_signal_aux
```

## Paso 4: Graficar el ECG con `ecg-plot`

```python
# Graficar el ECG usando ecg-plot
ecg_plot.plot_1(ecg_signal_aux)
plt.show()
```

```python
# Graficar el ECG usando ecg-plot
ecg_plot.plot_1(ecg_signal)
plt.show()
```

## Código Completo

```python
# Generar la señal base y crear derivaciones
base_ecg = nk.ecg_simulate(duration=10, sampling_rate=500)
#base_ecg = nk.ecg_simulate(duration=10, method="multileads")

ecg_leads = [base_ecg + (i * 0.01) for i in range(12)]

# Graficar las 12 derivaciones
ecg_plot.plot_12(ecg_leads)
plt.show()
```
